# 整合所有代码

```python
import pandas as pd
df = pd.read_excel("data\schedule.xls")

df.name = df.columns[0]

df.columns = ["col{i}".format(i=i) for i,x in enumerate(df.columns)]    #改数据标题名称为col0, col1, ....
df = df.fillna(value='')                                                #fillna

# each_profession_start 各专业起始行数
each_profession_start = { i:x for i,x in enumerate(df['col0'].values) if "院系：" in x }

# 修改数据标题
line = min(each_profession_start.keys())   # 第一个专业起始行数  min(最小值)
listA = [x.strip().replace(" ","") if len(x)>0 else '上课时间' for x in list(df.iloc[line+1])]
listB = [x.strip().replace(" ","") for x in list(df.iloc[line+2])]
listC = [x+listB[i] for i,x in enumerate(listA)]
df.columns = listC

df_output = pd.DataFrame()  #创新的空的输出结果DataFrame
# 每个专业从起始行数+3行开始，到下一个专业起始行数，一个个专业抽取数据并加上 院系	专业 两栏位
lines_start = sorted(list(each_profession_start.keys()))
for i in range(len(lines_start)-1):
    df_one_profession = df.iloc[lines_start[i]+3:lines_start[i+1]]  # df_one_profession = df.iloc[80+3:93] 

    #取 院系	专业 两栏位所需的值
    dict_add = {x.split("：")[0]:x.split("：")[1] for x in [x for x in df.iloc[lines_start[i]]['面向对象'].split(" ") if len(x)>0 ]}
    
    #加上 院系	专业 两栏位
    for k, v in dict_add.items():
        df_one_profession[k] = v

    #把一项专业的输出结果输出到DataFrame df_output
    df_output = pd.concat([df_output, df_one_profession] )   
    
df_output    
```
