```python
with open('text.csv', encoding='utf8') as fp:
    tt = fp.readlines()                    # 档案对象方法.readlines()

lines_appear = []                     #空列表，找到对的就放
for i, x in enumerate(tt):          #变数，一行行构建出的列表  BIF enumerate()
    if '院系：艺术设计与创意产业系' in x:                 #在一行内找到词的话
        lines_appear.append(i)        #加入行数
lines_appear        
```
